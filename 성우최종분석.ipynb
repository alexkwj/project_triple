### 이상치 처리

```python
import pandas as pd
import numpy as np

def find_outliers_iqr(df, columns):
    """
    IQR 방식을 사용하여 데이터프레임의 여러 열에 대한 이상치를 식별하는 함수.

    :param df: pandas.DataFrame - 이상치를 찾을 데이터프레임
    :param columns: list - 이상치를 검사할 열 이름 리스트
    :return: dict - 각 열의 이상치 값을 담고 있는 딕셔너리
    """
    outliers_dict = {}

    for col in columns:
        # 수치형 데이터가 아닌 경우 건너뜀
        if not np.issubdtype(df[col].dtype, np.number):
            print(f"'{col}' 열은 수치형이 아니므로 건너뜁니다.")
            continue

        # 1사분위수(Q1), 3사분위수(Q3) 계산
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # 이상치 경계 계산
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # 이상치 식별
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

        if not outliers.empty:
            outliers_dict[col] = outliers.tolist()

    return outliers_dict
```

```python
try:
    outlier_results = find_outliers_iqr(df, col)

    # 4. 결과 출력
    print("\n--- 이상치 탐지 결과 ---")
    if outlier_results:
        for column, outliers in outlier_results.items():
            # 이상치가 너무 많으면 개수와 일부만 출력
            if len(outliers) > 10:
                print(f"'{column}' 열에서 {len(outliers)}개의 이상치 발견. (예시: {outliers[:5]})")
            else:
                print(f"'{column}' 열의 이상치: {outliers}")
    else:
        print("지정된 열에서 이상치가 발견되지 않았습니다.")

except FileNotFoundError:
    print("final.csv 파일을 찾을 수 없습니다.")
```

```python
import pandas as pd
import numpy as np

def treat_outliers_iqr(df, columns):
    """
    IQR 방식을 사용하여 식별된 이상치를 상한/하한 값으로 대체(캡핑)하는 함수.
    """
    df_copy = df.copy()
    print("--- 이상치 처리 시작 ---")

    for col in columns:
        if not np.issubdtype(df_copy[col].dtype, np.number):
            print(f"'{col}' 열은 수치형이 아니므로 건너뜁니다.")
            continue

        # 결측치가 아닌 값들로만 계산
        valid_data = df_copy[col].dropna()
        if valid_data.empty:
            print(f"'{col}' 열에 유효한 데이터가 없어 건너뜁니다.")
            continue

        Q1 = valid_data.quantile(0.25)
        Q3 = valid_data.quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers_count = ((df_copy[col] < lower_bound) | (df_copy[col] > upper_bound)).sum()

        if outliers_count > 0:
            print(f"'{col}' 열: {outliers_count}개의 이상치를 경계 값으로 대체합니다.")
            df_copy[col] = df_copy[col].clip(lower_bound, upper_bound)
        else:
            print(f"'{col}' 열: 처리할 이상치가 없습니다.")

    print("--- 이상치 처리 완료 ---")
    return df_copy
```

```python
try:
    # 1. 처리할 열을 'INCOME', 'HOUSE_INCOME'으로만 지정
    columns_to_treat = ['INCOME', 'HOUSE_INCOME']

    # 2. 처리 전 통계 확인
    print("--- 처리 전 통계 ---")
    print(df[columns_to_treat].describe())

    # 3. 함수 실행
    df_income_treated = treat_outliers_iqr(df, columns_to_treat)

    # 4. 처리 후 통계 확인
    print("\n--- 처리 후 통계 ---")
    print(df_income_treated[columns_to_treat].describe())

except FileNotFoundError:
    print("final.csv 파일을 찾을 수 없습니다.")
```

#### 수정된 피쳐

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
# 1. 피처 선택
# 'HOUSE_INCOME' 제외하고 피처 재정의
numerical_features = [
    'AGE_GRP',
    'TRAVEL_STYL_1_GROUP', 'TRAVEL_STYL_2_GROUP', 'TRAVEL_STYL_3_GROUP',
    'TRAVEL_STYL_4_GROUP', 'TRAVEL_STYL_5_GROUP', 'TRAVEL_STYL_6_GROUP',
    'TRAVEL_STYL_7_GROUP', 'TRAVEL_STYL_8_GROUP'
]
categorical_features = ['TRAVEL_STATUS_ACCOMPANY']

features = numerical_features + categorical_features
df_selected = df[features].copy()

# 2. 전처리 파이프라인 구축
# 수치형 데이터 처리: 결측치는 중앙값으로 대체 후, 표준화(StandardScaler) 적용
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# 범주형 데이터 처리: 결측치는 최빈값으로 대체 후, 원-핫 인코딩(OneHotEncoder) 적용
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# 수치형과 범주형 파이프라인 통합
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# 3. 파이프라인 실행
X_processed = preprocessor.fit_transform(df_selected)

# 4. 결과 확인
print("--- 데이터 전처리 파이프라인 실행 완료 ---")
print("원본 데이터 형태 (선택된 피처):", df_selected.shape)
print("전처리 후 데이터 형태:", X_processed.shape)
```

```python
# 4. DataFrame으로 변환
new_column_names = preprocessor.get_feature_names_out()

# 주석: X_processed가 이미 일반 배열이므로 .toarray()를 삭제했어.
processed_df = pd.DataFrame(X_processed, columns=new_column_names)


# 5. CSV 파일로 저장
output_filename = 'final_processed_data.csv'
processed_df.to_csv(output_filename, index=False)



# 6. 결과 확인
# (이후 코드는 동일)
# ...
```

```python
import matplotlib.pyplot as plt
from koreanize_matplotlib import koreanize
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score
import numpy as np

# 한글 폰트 활성화
koreanize()

# --------- 데이터 설정 ---------
sample_size = 3000  # 14만 중 3000개 샘플링 (실루엣 계산용)
sample_idx = np.random.choice(X_processed.shape[0], sample_size, replace=False)
X_sample = X_processed[sample_idx]

K = range(2, 11)
inertias = []
sil_scores = []

for k in K:
    # MiniBatchKMeans (대용량에 적합)
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=2048)
    labels = kmeans.fit_predict(X_processed)

    inertias.append(kmeans.inertia_)

    # 실루엣 점수는 샘플링 데이터로 근사 계산
    sil = silhouette_score(X_sample, kmeans.predict(X_sample))
    sil_scores.append(sil)

# --------- 시각화 ---------
plt.figure(figsize=(12, 5))

# 엘보우 기법
plt.subplot(1, 2, 1)
plt.plot(K, inertias, "o-")
plt.xlabel("클러스터 개수 (k)")
plt.ylabel("관성값 (Inertia)")
plt.title("엘보우 기법 (MiniBatchKMeans)")

# 실루엣 기법
plt.subplot(1, 2, 2)
plt.plot(K, sil_scores, "o-")
plt.xlabel("클러스터 개수 (k)")
plt.ylabel("실루엣 점수 (샘플링)")
plt.title("실루엣 기법 (샘플링)")

plt.show()
```

### train,test 불러오기 , 전처리 한번에

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pickle

# --- 0. 파일 읽기 ---
# 1. Train 데이터 경로를 노트북에 있는 정확한 파일 이름으로 수정
train_df = pd.read_csv('/content/drive/MyDrive/pro5/project5/final.csv', low_memory=False)
test_df = pd.read_csv('/content/drive/MyDrive/pro5/project5/final_merged_test_data_ALL.csv', low_memory=False)


# --- 2. Test 데이터를 Train과 같게 (노트북의 맞춤 전처리 로직 적용) ---

def custom_preprocess(df):
    """
    siiuuuuuuu님의 노트북에 있던 전처리 과정을 수행하는 함수
    """
    # 주석: 필요한 21개 컬럼만 먼저 선택. TRAVELER_ID 등은 마지막에 제거.
    initial_features = [
        'TRAVELER_ID','GENDER','AGE_GRP','INCOME','MARR_STTS','HOUSE_INCOME',
        'TRAVEL_LIKE_SIDO_1', 'TRAVEL_STYL_1','TRAVEL_STYL_2','TRAVEL_STYL_3',
        'TRAVEL_STYL_4','TRAVEL_STYL_5','TRAVEL_STYL_6','TRAVEL_STYL_7','TRAVEL_STYL_8',
        'TRAVEL_STATUS_ACCOMPANY', 'VISIT_AREA_ID','VISIT_AREA_NM','LODGING_NM','TRAVEL_TERM'
    ]
    # 주석: 파일에 없는 컬럼이 있어도 오류나지 않도록 처리
    existing_initial_features = [col for col in initial_features if col in df.columns]
    df = df[existing_initial_features].copy()

    # HOUSE_INCOME 결측치를 중앙값으로 채우기
    df['HOUSE_INCOME'] = df['HOUSE_INCOME'].fillna(df['HOUSE_INCOME'].median())

    # TRAVEL_STYL_* 컬럼을 새로운 그룹으로 묶는 함수 정의
    def group_travel_style(x):
        if x in [1, 2]: return 1
        elif x == 3: return 2
        else: return 3

    # 8개의 TRAVEL_STYL 컬럼에 함수 적용 및 기존 컬럼 삭제
    for i in range(1, 9):
        style_col = f'TRAVEL_STYL_{i}'
        new_col = f'TRAVEL_STYL_{i}_GROUP'
        if style_col in df.columns:
            df[new_col] = df[style_col].apply(group_travel_style)
            df.drop(style_col, axis=1, inplace=True)

    # 범주형 변수 원-핫 인코딩
    df = pd.get_dummies(df, columns=['GENDER', 'TRAVEL_STATUS_ACCOMPANY'], dtype=float)

    # 불필요한 컬럼 최종 제거
    cols_to_drop = ['TRAVELER_ID', 'VISIT_AREA_ID', 'VISIT_AREA_NM', 'LODGING_NM']
    existing_cols_to_drop = [col for col in cols_to_drop if col in df.columns]
    df.drop(existing_cols_to_drop, axis=1, inplace=True)

    return df

print("--- 전처리 시작 ---")
X_train = custom_preprocess(train_df)
X_test = custom_preprocess(test_df)

# 원-핫 인코딩 후 Train/Test 컬럼 수가 다를 수 있으므로 맞춰주기
# 주석: Train에만 있는 컬럼은 Test에도 만들고, Test에만 있는 컬럼은 제거.
train_labels = X_train.columns
test_labels = X_test.columns

missing_in_test = set(train_labels) - set(test_labels)
for c in missing_in_test:
    X_test[c] = 0

missing_in_train = set(test_labels) - set(train_labels)
X_test.drop(list(missing_in_train), axis=1, inplace=True)

# 컬럼 순서도 동일하게 맞춰준다
X_test = X_test[train_labels]

print(f"Train 데이터 전처리 후 Shape: {X_train.shape}")
print(f"Test 데이터 전처리 후 Shape: {X_test.shape}")


# --- 3. 데이터 스케일링 ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) # 주석: 여긴 transform만!


# --- 4. 모델 학습 및 예측 ---
print("\n--- 모델 학습 및 예측 ---")
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
print("K-Means 모델 학습 시작...")
kmeans.fit(X_train_scaled)
print("모델 학습 완료.")

print("Test 데이터 예측 시작...")
test_cluster_labels = kmeans.predict(X_test_scaled)
print("Test 데이터 예측 완료.")


# --- 5. 추가 작업: 모델, 스케일러, 결과 저장 ---
print("\n--- 추가 작업 수행 ---")

# 스케일러 저장
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
print("스케일러가 'scaler.pkl' 파일로 저장되었습니다.")

# K-Means 모델 저장
with open('kmeans_model.pkl', 'wb') as f:
    pickle.dump(kmeans, f)
print("K-Means 모델이 'kmeans_model.pkl' 파일로 저장되었습니다.")

# 최종 예측 결과를 원본 Test 데이터프레임에 추가
test_df['predicted_cluster'] = test_cluster_labels

# 결과 CSV 파일로 저장
output_filename = 'test_data_with_predictions.csv'
test_df.to_csv(output_filename, index=False, encoding='utf-8-sig')
print(f"최종 예측 결과가 '{output_filename}' 파일로 저장되었습니다.")

# Test 데이터의 군집별 분포 확인
print("\n--- 최종 예측 결과 확인 ---")
print("Test 데이터의 군집별 분포:")
print(test_df['predicted_cluster'].value_counts().sort_index())
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# (사전 준비 코드는 동일)
# ...

# 1. 원본 데이터에 학습된 군집 라벨(cluster) 추가
train_df['cluster'] = kmeans.labels_

# 2. (1단계) 의미 없는 장소(노이즈) 필터링
# 주석: 추천에 방해되는 공통 장소들을 제거 리스트로 만들어.
stopwords = [
    # 기존 목록
    '제주 국제공항', '김포국제공항 국내선', '사무실', '제주동문시장',
    '서귀포 매일 올레시장', '서울역', '본점', '스타벅스', 'CU', 'GS25',
    '함덕해수욕장', '협재해수욕장', '청주국제공항', '김포 국제공항 국내선','성산일출봉','오설록 티 뮤지엄', '성심당 본점',

    # 새로 추가된 역/공항/터미널
    '부산역', '대전역', '광주공항', '서울고속버스터미널','서울고속버스터미널'

    # 그 외 추가하면 좋을만한 주요 교통 허브
    '동대구역', '인천국제공항', '광명역', '울산역','경부','경부'
]

# 주석: 노이즈가 포함된 행을 제외한 새로운 데이터프레임을 만든다.
filtered_df = train_df[~train_df['VISIT_AREA_NM'].isin(stopwords)].copy()


# 3. (2단계) 필터링된 데이터로 TF-IDF 실행
# 주석: 이제부터는 train_df가 아닌 filtered_df를 사용해.
filtered_df['VISIT_AREA_NM_CLEAN'] = filtered_df['VISIT_AREA_NM'].str.replace(' ', '_', regex=False).fillna('')

corpus = filtered_df.groupby('cluster')['VISIT_AREA_NM_CLEAN'].apply(
    lambda x: ' '.join(x)
).tolist()

vectorizer = TfidfVectorizer(min_df=2)
tfidf_matrix = vectorizer.fit_transform(corpus)
feature_names = np.array(vectorizer.get_feature_names_out())

# 4. 최종 추천 프로필 생성
recommendation_profile_best = {}
for i, doc_vector in enumerate(tfidf_matrix):
    top_indices = doc_vector.toarray()[0].argsort()[-5:][::-1]
    top_features = [name.replace('_', ' ') for name in feature_names[top_indices]]
    recommendation_profile_best[i] = top_features

# 5. 최종 결과 확인
print("\n--- [최종 개선] 필터링 + TF-IDF 적용 후 추천 프로필 ---")
for cluster, places in recommendation_profile_best.items():
    print(f"Cluster {cluster}: {places}")
```

#### 1. 예측 파이프라인 함수 코드


```python
# 기존 코드의 전처리 파이프라인 정의 부분 (셀 번호 61)

import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
import pickle

# --- 1. 데이터 불러오기 및 피처 선택 ---
df = pd.read_csv('/content/drive/MyDrive/pro5/project5/final.csv', low_memory=False)

numerical_features = [
    'AGE_GRP',
    'TRAVEL_STYL_1_GROUP', 'TRAVEL_STYL_2_GROUP', 'TRAVEL_STYL_3_GROUP',
    'TRAVEL_STYL_4_GROUP', 'TRAVEL_STYL_5_GROUP', 'TRAVEL_STYL_6_GROUP',
    'TRAVEL_STYL_7_GROUP', 'TRAVEL_STYL_8_GROUP'
]

# 주석: RESIDENCE_SGG_CD를 범주형 피처에 추가한다
categorical_features = ['TRAVEL_STATUS_ACCOMPANY', 'RESIDENCE_SGG_CD']

features = numerical_features + categorical_features
# 주석: 데이터프레임에서 필요한 피처만 선택한다
df_selected = df[features].copy()

# --- 2. 전처리 파이프라인 정의 (이후 코드는 동일) ---
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    # 주석: 원-핫 인코딩 시 알려지지 않은 카테고리는 무시한다
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# --- 3. 전처리 및 모델 학습 ---
print("--- 전처리 및 모델 학습 시작 ---")
X_processed = preprocessor.fit_transform(df_selected)
# 주석: K-means 모델을 4개 군집으로 학습시킨다
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
kmeans.fit(X_processed)
print("--- 학습 완료 ---")

# --- 4. 학습된 전처리기와 모델 저장 ---
with open('preprocessor_final_with_residence.pkl', 'wb') as f:
    pickle.dump(preprocessor, f)
print("전처리기가 'preprocessor_final_with_residence.pkl' 파일로 저장되었습니다.")

with open('kmeans_model_final_with_residence.pkl', 'wb') as f:
    pickle.dump(kmeans, f)
print("K-Means 모델이 'kmeans_model_final_with_residence.pkl' 파일로 저장되었습니다.")
```

```python
# 기존 코드의 신규 고객 정보 입력 부분 (셀 번호 71)

import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer

# --- 1. 필요한 모든 파일과 정보 미리 불러오기 ---
try:
    # 주석: 새로운 전처리기와 모델 파일을 로딩한다
    with open('preprocessor_final_with_residence.pkl', 'rb') as f:
        preprocessor = pickle.load(f)
    with open('kmeans_model_final_with_residence.pkl', 'rb') as f:
        kmeans = pickle.load(f)
    train_df = pd.read_csv('/content/drive/MyDrive/pro5/project5/final.csv', low_memory=False)
    print("--- 모든 파일 로딩 완료 ---")

    numerical_features = [
        'AGE_GRP',
        'TRAVEL_STYL_1_GROUP', 'TRAVEL_STYL_2_GROUP', 'TRAVEL_STYL_3_GROUP',
        'TRAVEL_STYL_4_GROUP', 'TRAVEL_STYL_5_GROUP', 'TRAVEL_STYL_6_GROUP',
        'TRAVEL_STYL_7_GROUP', 'TRAVEL_STYL_8_GROUP'
    ]
    # 주석: RESIDENCE_SGG_CD를 범주형 피처에 추가한다
    categorical_features = ['TRAVEL_STATUS_ACCOMPANY', 'RESIDENCE_SGG_CD']
    features = numerical_features + categorical_features

    df_selected_train = train_df[features].copy()
    X_processed_train = preprocessor.transform(df_selected_train)

    train_cluster_labels = kmeans.predict(X_processed_train)
    train_df['cluster'] = train_cluster_labels

    # 추천 프로필 생성 (기존 코드와 동일)
    stopwords = [
    '제주 국제공항', '김포국제공항 국내선', '사무실', '제주동문시장',
    '서귀포 매일 올레시장', '서울역', '본점', '스타벅스', 'CU', 'GS25',
    '함덕해수욕장', '협재해수욕장', '청주국제공항', '김포 국제공항 국내선','성산일출봉','오설록 티 뮤지엄', '성심당 본점',
    '부산역', '대전역', '광주공항', '서울고속버스터미널','서울고속버스터미널',
    '동대구역', '인천국제공항', '광명역', '울산역','경부','경부'
    ]
    filtered_df = train_df[~train_df['VISIT_AREA_NM'].isin(stopwords)].copy()
    filtered_df['VISIT_AREA_NM_CLEAN'] = filtered_df['VISIT_AREA_NM'].str.replace(' ', '_', regex=False).fillna('')
    corpus = filtered_df.groupby('cluster')['VISIT_AREA_NM_CLEAN'].apply(lambda x: ' '.join(x)).tolist()
    vectorizer = TfidfVectorizer(min_df=2)
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = np.array(vectorizer.get_feature_names_out())
    recommendation_profile = {}
    for i, doc_vector in enumerate(tfidf_matrix):
        top_indices = doc_vector.toarray()[0].argsort()[-5:][::-1]
        top_features = [name.replace('_', ' ') for name in feature_names[top_indices]]
        recommendation_profile[i] = top_features
    print("--- 군집별 추천 프로필 생성 완료 ---")

except FileNotFoundError as e:
    print(f"오류: '{e.filename}' 파일이 없습니다. 필요한 파일들을 확인해주세요.")
    exit()

# --- 2. 예측 및 추천 함수 정의 ---
def predict_cluster(new_customer_data):
    df_new = pd.DataFrame([new_customer_data])
    # 주석: 예측을 위한 피처 목록에 RESIDENCE_SGG_CD를 추가한다
    expected_features = numerical_features + categorical_features
    df_new_aligned = df_new.reindex(columns=expected_features)
    X_processed_new = preprocessor.transform(df_new_aligned)
    predicted_cluster = kmeans.predict(X_processed_new)
    return int(predicted_cluster[0])

def get_recommendations(cluster_id):
    return recommendation_profile.get(cluster_id, ["추천 목록을 찾을 수 없습니다."])

# --- 3. 신규 고객 정보 입력받기 (수정) ---
print("\n--- 신규 고객 정보 입력을 시작합니다. ---")
new_user_input = {}

style_questions = {
    'TRAVEL_STYL_1': {1: '자연', 2: '상관없음', 3: '도시'},
    'TRAVEL_STYL_2': {1: '숙박', 2: '상관없음', 3: '당일치기'},
    'TRAVEL_STYL_3': {1: '새로운지역', 2: '상관없음', 3: '익숙한지역'},
    'TRAVEL_STYL_4': {1: '편하지만비싼숙소', 2: '상관없음', 3: '불편하지만저렴함숙소'},
    'TRAVEL_STYL_5': {1: '휴양/휴식', 2: '상관없음', 3: '엑티비티'},
    'TRAVEL_STYL_6': {1: '숨은명소', 2: '상관없음', 3: '유명한명소'},
    'TRAVEL_STYL_7': {1: '계획여행', 2: '상관없음', 3: '즉흥여행'},
    'TRAVEL_STYL_8': {1: '사진촬영선호', 2: '상관없음', 3: '사진촬영비선호'}
}
accompany_options = {
    1: '나홀로 여행', 2: '자녀 동반 여행', 3: '2인 가족 여행', 4: '2인 여행(가족 외)',
    5: '부모 동반 여행', 6: '3인 이상 여행(가족 외)', 7: '3대 동반 여행(친척 포함)', 8: '3인 이상 가족 여행(친척 포함)'
}

# 주석: 행정번호-지역명 맵핑을 위한 딕셔너리를 생성한다
residence_map = {
    41: '경기도', 11: '서울특별시', 28: '인천광역시', 48: '경상남도', 45: '전라북도',
    44: '충청남도', 50: '제주특별자치도', 27: '대구광역시', 26: '부산광역시', 30: '대전광역시',
    36: '세종특별자치시', 29: '광주광역시', 31: '울산광역시', 47: '경상북도', 46: '전라남도',
    43: '충청북도', 42: '강원도'
}
residence_options_rev = {v: k for k, v in residence_map.items()}

try:
    print("\n1. 가고싶은 지역을 선택하세요:")
    for i, (name, code) in enumerate(residence_options_rev.items(), 1):
        print(f"   {i}: {name}")
    choice_residence = int(input("   선택 (번호 입력): "))
    selected_name = list(residence_options_rev.keys())[choice_residence - 1]
    new_user_input['RESIDENCE_SGG_CD'] = residence_options_rev[selected_name]

    new_user_input['AGE_GRP'] = int(input("2. 연령대를 입력하세요 (예: 20, 30, 40): "))

    # 주석: 거주지 행정번호를 한글명으로 선택하게 한다

    print("\n3. 동반 여행 형태를 선택하세요:")
    for key, value in accompany_options.items():
        print(f"   {key}: {value}")
    choice = int(input("   선택 (숫자 입력): "))
    if choice in accompany_options:
        new_user_input['TRAVEL_STATUS_ACCOMPANY'] = accompany_options[choice]
    else:
        raise ValueError("목록에 없는 번호입니다.")

    print("\n여행 스타일에 대해 해당하는 번호를 선택해주세요.")
    for i in range(1, 9):
        style_key = f'TRAVEL_STYL_{i}'
        options = style_questions[style_key]
        options_text = ", ".join([f"{key}: {value}" for key, value in options.items()])
        user_choice = int(input(f" - 스타일 {i} ({options_text}): "))
        if user_choice in options:
            new_user_input[f'TRAVEL_STYL_{i}_GROUP'] = user_choice
        else:
            raise ValueError("목록에 없는 번호입니다.")

    # --- 4. 최종 추천 시스템 실행 ---
    predicted_cluster_id = predict_cluster(new_user_input)
    recommended_places = get_recommendations(predicted_cluster_id)

    cluster_descriptions = {
        0: "당신은 자연 속에서 여유롭게 쉬는 여행을 선호하는  여행자군요 🧘‍♀️",
        1: "당신은 가족을 좋아하며 동반 여행을 선호하는 여행자군요 👨‍👩‍👧‍👦",
        2: "당신은 체험을 중시하며 거의 모든 활동을 즐기는 친구·연인 동반을 선호하는 만능 액티비티 여행자군요 🧗‍♀️",
        3: "당신은 음식과 자연에 집중하며 친구·연인 또는 혼자 여행을 선호하는 가성비 미식·탐험 여행자군요 🍜"
    }

    print(f"\n================ 최종 추천 결과 ================")
    print(f"입력하신 고객님은 [ {predicted_cluster_id}번 군집 ] 유형입니다.")
    print(cluster_descriptions.get(predicted_cluster_id, "군집 설명을 찾을 수 없습니다."))
    print(f"당신과 같은 사람이 자주 간 여행지는 다음과 같습니다 : {recommended_places}")

except (ValueError, IndexError) as e:
    print(f"\n오류: 잘못된 값이 입력되었습니다. {e}")
except Exception as e:
    print(f"\n알 수 없는 오류가 발생했습니다: {e}")
```

```python
import streamlit as st
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans

# --- 1. 데이터 및 모델 로딩 (로컬 파일 필요) ---
@st.cache_resource
def load_resources():
    """
    전처리기, 모델, 학습 데이터를 로드하는 함수.
    로컬 파일이 없으면 임시로 학습을 진행한다.
    """
    # 주석: features 변수를 함수 시작 부분에 선언한다.
    numerical_features = [
        'AGE_GRP', 'TRAVEL_STYL_1_GROUP', 'TRAVEL_STYL_2_GROUP', 'TRAVEL_STYL_3_GROUP',
        'TRAVEL_STYL_4_GROUP', 'TRAVEL_STYL_5_GROUP', 'TRAVEL_STYL_6_GROUP',
        'TRAVEL_STYL_7_GROUP', 'TRAVEL_STYL_8_GROUP'
    ]
    categorical_features = ['TRAVEL_STATUS_ACCOMPANY', 'RESIDENCE_SGG_CD']
    features = numerical_features + categorical_features

    preprocessor = None
    kmeans = None
    train_df = None

    try:
        # 주석: 실제 파일을 불러오는 코드.
        with open('preprocessor_final_with_residence.pkl', 'rb') as f:
            preprocessor = pickle.load(f)
        with open('kmeans_model_final_with_residence.pkl', 'rb') as f:
            kmeans = pickle.load(f)
        train_df = pd.read_csv('/content/drive/MyDrive/pro5/project5/final.csv', low_memory=False)
        st.success("모델과 데이터를 성공적으로 로드했어!")

    except FileNotFoundError:
        st.warning("경고: 로컬 모델 파일이 없어 임시로 학습을 진행한다. 시간이 다소 소요될 거야.")
        train_df = pd.read_csv('/content/drive/MyDrive/pro5/project5/final.csv', low_memory=False)

        # 주석: 전처리 파이프라인과 모델을 학습시킨다.
        df_selected_train = train_df[features].copy()

        numeric_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ])
        categorical_transformer = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ])
        preprocessor = ColumnTransformer(
            transformers=[
                ('num', numeric_transformer, numerical_features),
                ('cat', categorical_transformer, categorical_features)
            ])

        X_processed_train = preprocessor.fit_transform(df_selected_train)
        kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
        kmeans.fit(X_processed_train)

    # 군집 라벨을 데이터프레임에 추가한다.
    train_df['cluster'] = kmeans.predict(preprocessor.transform(train_df[features]))

    # TF-IDF 기반 추천 프로필 생성
    stopwords = [
        '제주 국제공항', '김포국제공항 국내선', '사무실', '제주동문시장',
        '서귀포 매일 올레시장', '서울역', '본점', '스타벅스', 'CU', 'GS25',
        '함덕해수욕장', '협재해수욕장', '청주국제공항', '김포 국제공항 국내선', '성산일출봉', '오설록 티 뮤지엄', '성심당 본점',
        '부산역', '대전역', '광주공항', '서울고속버스터미널', '서울고속버스터미널',
        '동대구역', '인천국제공항', '광명역', '울산역', '경부', '경부'
    ]
    filtered_df = train_df[~train_df['VISIT_AREA_NM'].isin(stopwords)].copy()
    filtered_df['VISIT_AREA_NM_CLEAN'] = filtered_df['VISIT_AREA_NM'].str.replace(' ', '_', regex=False).fillna('')
    corpus = filtered_df.groupby('cluster')['VISIT_AREA_NM_CLEAN'].apply(lambda x: ' '.join(x)).tolist()
    vectorizer = TfidfVectorizer(min_df=2)
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = np.array(vectorizer.get_feature_names_out())

    recommendation_profile = {}
    for i, doc_vector in enumerate(tfidf_matrix):
        top_indices = doc_vector.toarray()[0].argsort()[-5:][::-1]
        top_features = [name.replace('_', ' ') for name in feature_names[top_indices]]
        recommendation_profile[i] = top_features

    return preprocessor, kmeans, recommendation_profile

# --- 2. 예측 및 추천 함수 정의 ---
def predict_cluster(preprocessor, kmeans, new_customer_data):
    numerical_features = [
        'AGE_GRP', 'TRAVEL_STYL_1_GROUP', 'TRAVEL_STYL_2_GROUP', 'TRAVEL_STYL_3_GROUP',
        'TRAVEL_STYL_4_GROUP', 'TRAVEL_STYL_5_GROUP', 'TRAVEL_STYL_6_GROUP',
        'TRAVEL_STYL_7_GROUP', 'TRAVEL_STYL_8_GROUP'
    ]
    categorical_features = ['TRAVEL_STATUS_ACCOMPANY', 'RESIDENCE_SGG_CD']
    features = numerical_features + categorical_features

    df_new = pd.DataFrame([new_customer_data])
    df_new_aligned = df_new.reindex(columns=features)
    X_processed_new = preprocessor.transform(df_new_aligned)
    predicted_cluster = kmeans.predict(X_processed_new)
    return int(predicted_cluster[0])

# --- 3. Streamlit 앱 인터페이스 구축 ---
st.set_page_config(page_title="여행지 추천 시스템", layout="wide")
st.title("나에게 딱 맞는 여행지 추천 시스템 ✈️")
st.markdown("당신의 여행 스타일을 알려주면, 맞춤형 여행지를 추천해 줄게!")

# 리소스 로딩
preprocessor, kmeans, recommendation_profile = load_resources()

# 군집 설명
cluster_descriptions = {
    0: "휴양과 여유로운 일정을 선호하는 웰니스 중심의 개인 여행자구나 🧘‍♀️",
    1: "다양한 활동에 적극적이며 동반 여행을 선호하는 여행자구나 👨‍👩‍👧‍👦",
    2: "체험을 중시하며 거의 모든 활동을 즐기는 친구·연인 동반을 선호하는 만능 액티비티 여행자구나 🧗‍♀️",
    3: "음식과 자연에 집중하며 친구·연인 또는 혼자 여행을 선호하는 가성비 미식·탐험 여행자구나 🍜"
}

# 사용자 입력 위젯
with st.form("user_input_form"):
    st.header("1. 기본 정보")
    age_grp = st.slider("연령대를 선택해줘:", min_value=10, max_value=80, value=30, step=10)

    residence_map = {
        '경기도': 41, '서울특별시': 11, '인천광역시': 28, '경상남도': 48, '전라북도': 45,
        '충청남도': 44, '제주특별자치도': 50, '대구광역시': 27, '부산광역시': 26, '대전광역시': 30,
        '세종특별자치시': 36, '광주광역시': 29, '울산광역시': 31, '경상북도': 47, '전라남도': 46,
        '충청북도': 43, '강원도': 42
    }
    selected_residence_name = st.selectbox("거주지를 선택해줘:", list(residence_map.keys()))
    residence_code = residence_map[selected_residence_name]

    accompany_options = {
        '나홀로 여행': '나홀로 여행', '자녀 동반 여행': '자녀 동반 여행', '2인 가족 여행': '2인 가족 여행', '2인 여행(가족 외)': '2인 여행(가족 외)',
        '부모 동반 여행': '부모 동반 여행', '3인 이상 여행(가족 외)': '3인 이상 여행(가족 외)', '3대 동반 여행(친척 포함)': '3대 동반 여행(친척 포함)', '3인 이상 가족 여행(친척 포함)': '3인 이상 가족 여행(친척 포함)'
    }
    selected_accompany = st.selectbox("동반 여행 형태를 선택해줘:", list(accompany_options.keys()))
    accompany_status = selected_accompany

    st.header("2. 여행 스타일")

    style_questions = {
        'TRAVEL_STYL_1_GROUP': '자연(1) vs 도시(3)',
        'TRAVEL_STYL_2_GROUP': '숙박(1) vs 당일치기(3)',
        'TRAVEL_STYL_3_GROUP': '새로운 지역(1) vs 익숙한 지역(3)',
        'TRAVEL_STYL_4_GROUP': '편하고 비싼 숙소(1) vs 저렴한 숙소(3)',
        'TRAVEL_STYL_5_GROUP': '휴양/휴식(1) vs 액티비티(3)',
        'TRAVEL_STYL_6_GROUP': '숨은 명소(1) vs 유명한 명소(3)',
        'TRAVEL_STYL_7_GROUP': '계획 여행(1) vs 즉흥 여행(3)',
        'TRAVEL_STYL_8_GROUP': '사진 촬영 선호(1) vs 사진 촬영 비선호(3)'
    }

    styles = {}
    for key, question in style_questions.items():
        styles[key] = st.slider(f"스타일 ({question})", 1, 3, 2)

    submitted = st.form_submit_button("추천받기")

if submitted:
    new_user_input = {
        'AGE_GRP': age_grp,
        'RESIDENCE_SGG_CD': residence_code,
        'TRAVEL_STATUS_ACCOMPANY': accompany_status,
    }
    new_user_input.update(styles)

    predicted_cluster_id = predict_cluster(preprocessor, kmeans, new_user_input)
    recommended_places = recommendation_profile.get(predicted_cluster_id, ["추천 목록을 찾을 수 없습니다."])

    st.success("추천 완료!")
    st.header("⭐ 당신을 위한 맞춤 여행지 ⭐")
    st.markdown(f"**당신은 {predicted_cluster_id}번 군집 유형에 속하며,**")
    st.markdown(f"**{cluster_descriptions.get(predicted_cluster_id, '군집 설명을 찾을 수 없다.')}**")
    st.markdown("---")
    st.subheader("추천 여행지:")
    for place in recommended_places:
        st.write(f"- {place}")
```
