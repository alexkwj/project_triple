# Task
Create a Python script that takes user input for Gender, Age Group, and travel style preferences (TRAVEL_STYL_1 to TRAVEL_STYL_8) with specific validation rules and textual descriptions for the travel styles. Use this input to filter the `woojin_select` DataFrame (which has been previously processed and saved to '/content/drive/MyDrive/woojin_select.csv') and display the matching `TRAVEL_ID`s. If no matches are found, inform the user.

## Define input prompts and validation

### Subtask:
Create clear prompts for each input required from the user (Gender, Age Group, TRAVEL_STYL_1 to TRAVEL_STYL_8). Define the validation rules for each input based on the user's specifications (e.g., Gender must be '남' or '여', Age Group must be multiples of 10, TRAVEL_STYL values must be 1, 2, or 3 with corresponding meanings).


**Reasoning**:
Define the textual descriptions for the recoded travel styles and create user-friendly prompts with validation logic for gender, age group, and travel styles.



```python
travel_style_descriptions = {
    'TRAVEL_STYL_1': {1: '자연', 2: '상관없음', 3: '도시'},
    'TRAVEL_STYL_2': {1: '숙박', 2: '상관없음', 3: '당일치기'},
    'TRAVEL_STYL_3': {1: '새로운지역', 2: '상관없음', 3: '익숙한지역'},
    'TRAVEL_STYL_4': {1: '편하지만비싼숙소', 2: '상관없음', 3: '불편하지만저렴함숙소'},
    'TRAVEL_STYL_5': {1: '휴양휴식', 2: '상관없음', 3: '엑티비티'},
    'TRAVEL_STYL_6': {1: '숨은명소', 2: '상관없음', 3: '유명한명소'},
    'TRAVEL_STYL_7': {1: '계획여행', 2: '상관없음', 3: '즉흥여행'},
    'TRAVEL_STYL_8': {1: '포토광', 2: '상관없음', 3: '절대안찍음'}
}

def get_gender_input():
    while True:
        gender = input("성별을 입력하세요 (남 또는 여): ").strip()
        if gender in ['남', '여']:
            return gender
        else:
            print("잘못된 입력입니다. '남' 또는 '여'로 입력해주세요.")

def get_age_group_input():
    while True:
        try:
            age_group = int(input("나이대를 입력하세요 (10, 20, 30, ...): "))
            if age_group > 0 and age_group % 10 == 0:
                return age_group
            else:
                print("잘못된 입력입니다. 10의 배수로 입력해주세요.")
        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

def get_travel_style_input(style_name):
    while True:
        description = travel_style_descriptions[style_name]
        prompt = f"선호도를 입력하세요 ({', '.join([f'{v}({k})' for k, v in description.items()])}): "
        user_input = input(prompt).strip()
        for key, value in description.items():
            if user_input == value or user_input == str(key):
                return key
        print("잘못된 입력입니다. 유효한 옵션 중 하나를 선택해주세요.")
```

## Implement input collection and validation

### Subtask:
Collect input from the user for each required field (Gender, Age Group, TRAVEL_STYL_1 to TRAVEL_STYL_8) and validate the input against the defined rules. If the input is invalid, prompt the user to enter the information again.


**Reasoning**:
Collect and validate user input for Gender, Age Group, and Travel Styles using the previously defined functions.



```python
gender_input = get_gender_input()
age_group_input = get_age_group_input()

travel_style_cols = [f'TRAVEL_STYL_{i}' for i in range(1, 9)]
travel_style_inputs = {}
for style_name in travel_style_cols:
    travel_style_inputs[style_name] = get_travel_style_input(style_name)

print("\n입력된 정보:")
print(f"성별: {gender_input}")
print(f"나이대: {age_group_input}")

print("여행 스타일 선호도:")
for style, value in travel_style_inputs.items():
    print(f"  {style}: {value} ({travel_style_descriptions[style][value]})")
```

## Filter dataframe based on inputs

### Subtask:
Filter the `woojin_select` DataFrame based on the validated user inputs for Gender, Age Group, and the recoded TRAVEL_STYL columns.


**Reasoning**:
Filter the DataFrame based on gender, age group, and the recoded travel style preferences.



```python
filtered_df = woojin_select[
    (woojin_select['GENDER'] == gender_input) &
    (woojin_select['AGE_GRP'] == age_group_input)
]

for style_name, preference in travel_style_inputs.items():
    filtered_df = filtered_df[filtered_df[style_name] == preference]

display(filtered_df)
```

## Extract and display travel id

### Subtask:
Extract the unique `TRAVEL_ID`s from the filtered DataFrame and display them to the user.


**Reasoning**:
Check if the filtered DataFrame is empty and then extract and display the unique TRAVEL_IDs if it's not empty.



```python
if filtered_df.empty:
    print("입력하신 조건에 맞는 여행 ID가 없습니다.")
else:
    unique_travel_ids = filtered_df['TRAVEL_ID'].unique()
    print("입력하신 조건에 맞는 여행 ID 목록:")
    for travel_id in unique_travel_ids:
        print(travel_id)
```

## Handle no matching ids

### Subtask:
If no `TRAVEL_ID`s match the given criteria, inform the user that no results were found and perhaps suggest broadening their criteria.


**Reasoning**:
Check if the filtered_df is empty and print a message if it is, suggesting broadening the criteria.



```python
if filtered_df.empty:
    print("입력하신 조건에 맞는 여행 ID가 없습니다. 더 넓은 조건으로 검색해보세요.")
```
